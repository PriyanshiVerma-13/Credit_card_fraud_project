Importing the Dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset to a pandas DataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset informations
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
#dropping missing values
credit_card_data=credit_card_data.dropna()
```

```python
#rechecking missing values
credit_card_data.isnull().sum()
```

```python
# distribution of legit and fraudulent txn
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced

0-->Legit Transaction

1-->Fraudulent Transaction

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class==0.0]
fraud = credit_card_data[credit_card_data.Class==1.0]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values of both these txns
credit_card_data.groupby('Class').mean()
```

Under-Sampling

Bulid a sample dataset containing similar legit txns and fraudulent txns

Number of Fradulent tnxs-->360

```python
legit_sample = legit.sample(n=360)
```

Concatenating two Dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data and Testing data

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y, test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

Model Traning (Logistic Regression)

```python
model=LogisticRegression(solver='lbfgs', max_iter=5000)
```

```python
# training the logistic regression model with training data
model.fit(X_train,Y_train)
```

Evaluation of Model based on Accuracy score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data:",training_data_accuracy)
```

```python
#accuracy on test_data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data:', test_data_accuracy)
```
